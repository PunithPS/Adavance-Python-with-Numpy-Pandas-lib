```python
#Q8. How many distinct `work_type`  are present in the data?
dist_work_type = df['work_type'].nunique()
print(dist_work_type)
```

```python
#Q9. Print the value counts of gender.
gender_cnt = df['gender'].value_counts()
print(gender_cnt)
```

```python
#Q10. Print the average avg_glucose_level for each gender.
df['avg_glucose_level'] = df['avg_glucose_level'].astype(float)
gender_gp = df.groupby('gender')

avg_glu_level = gender_gp['avg_glucose_level'].mean()

print(avg_glu_level)
```
